## Introduction

When simulating the time-varying (ground)water table, it may cross the elevation of the tile drains or, even, ground surface. When this happens de drainage resistance changes from $\infty$ to $c_{dr}$ when the head rises or from $c_{dr}$ to $\infty$ when the head is descending.
To deal with this in a managed fashion whenever this happens during a time step, the moment at which the water table hits the drainage level is computed. The water table is first simulated up to the point at which it touches the drain elevation. Once the resistance is changed, the simulation is resumed until the end of the timestep. The subject is to define and implement a stable simulation without sawtoothing of the water table about the drainage level.

This notebook shows, that it is possible that a rising water table turn to a declining one upon passing the drainage level, where the drainage resistance changes from infinity to a low value. This can not in general be prevented, and, therefore, the swawtoothing issue is not solved by an implementation that simulates onto the time that the drainage level is passed and then simply contines with the adapted drainage resistance. This sawtoothin, initially unexpected was was difficult to prove. However it is now shown in this notebook. Whether this happens depends on the combination of the drainage elevation, the ditch level and the head in the underlying aquifer and teh value of $c_{dr}. Therfore, in general this method is not suitable.

Instead a more robust method was chosen, which is also demonstrated in this notebook. It too computes the moment at which the water table hits the drainage level. But from that moment on, it subtract the amount of drainage from the recharge. This amount is the exact drainage that occurs with drainage resistance equal to 0. The result is that the water table is constant and equal to the drainage level for times larger than the time of the hit.

# General approach of simulation with drainage

The governing partial differential equation

$$kD\frac{d^{2}h}{dx^{2}}	=-N+\frac{h-\phi}{c}+\frac{h-z_{dr}}{c_{dr}}$$

is equivalent to 

$$kD\frac{d^{2}h}{dx^{2}}=-N+\frac{h-\hat{\phi}}{\hat{c}}$$

with

$$\hat{c}=\frac{c_{dr}c}{c_{dr}+c}$$

and

$$\hat{\phi}=\frac{c_{dr}\phi+cz_{dr}}{c_{dr}+c}$$

Therefore, the solution for the transient approach followd in GGOR is

$$\overline{h}-\widehat{\phi}=\left(\overline{h}_{0}-\widehat{\phi}\right)e^{-\frac{t-t_{0}}{\widehat{T}}}+\left\{ N\widehat{c}\left(1 - \hat{\Lambda}\right)-\widehat{\Lambda}\left(h_{LR}-\widehat{\phi}\right)\right\} \left(1-e^{-\frac{t-t_{0}}{\widehat{T}}}\right)$$

with

$$\widehat{T}=\mu\widehat{c}$$

and

$$\widehat{\Lambda}=\frac{1}{\frac{b}{\widehat{\lambda}}\mbox{ctanh}\frac{b}{\widehat{\lambda}}+\frac{b/\widehat{c}}{D/w}}$$

To illustrate our point, we simulate the a rising head that passes the drainage level and then switches the drainage resistance from $\infty$ to some small value and see what happens and how this relates to the underlying theory.

For this we'll compute the moment at which the head passes the drainage level first, after which simulation goes on with the lower drainage resistance.

We'll also demonstrate the alternative, which consists of simulating until the drainage level is hit and then subtract exactly the amount of water that will keep the water table at the drainage level for the rest of the time. This amount of water is the drainage flux or the surface runoff, depending on which intepretation is given to the drainage level.

## Finding the moment at which the drainage level is hit

We have, assuming for convenience that $t_0 = 0$

$$\overline h - \phi = (\overline h_0 - \phi) e^{-\frac{t}{\hat T}} + s_\infty \left(1 - e^{-\frac{t}{\hat T}}\right)$$

where

$$s_\infty = N \hat{c} \left(1 - \hat{\Lambda}\right) - \left(h_{LR} - \hat{\phi}\right) \hat{\Lambda}$$

so that

$$\frac{d \overline h}{dt} =  \frac 1 {\hat T} \left( -\overline h_0 + \hat \phi + s_{\infty}\right) e^{-\frac t T} $$

The water table rises when

$$s_\infty + \hat \phi > \overline h_0$$

The moment at which the water table will intersect the drainage level $z_{dr}$ follows from

$$z_{dr} - \hat\phi = \left(\overline h_0 - \hat\phi\right) e^{-\frac t {\hat T}}
+ s_\infty \left(1 - e^{-\frac t {\hat\phi}}\right)$$

which yields

$$\tau = \hat T \ln\left(\frac{\overline h_0 - \hat\phi - s_\infty}
{z_{dr} - \hat\phi - s_\infty}\right)$$

This only yields an intersetion in the future when the argument of the logarithm is larger than 1. In the model it is, therefore, easily checked whether the intersction falls within the current time step.

$$r = \frac{\overline h_0 - \hat\phi - s_\infty}{z_{dr} - \hat\phi - s_\infty} > 1?$$

And if this is the case, then move on to the intersection time and, change the drainage resistance and move on.

# Implementation of the model

```python
import matplotlib.pyplot as plt
import numpy as np
#from scipy import special
import pandas as pd
import pdb

import os
import sys

HOME = os.path.expanduser('~/GRWMODELS/python/GGOR')
sys.path.insert(0, os.path.join(HOME, 'src/analytic'))

import ggor_analytical as ga
```

```python
def get_hatpars(b=None, mu=None, c=None, k=None, D=None,
                h0=None, phi=None, hdr=None, cdr=None,
                w=None, hlr=None, N=None, **kw):
    """Return the parameters with hat in the drainage formulas"""
    ch   = c / (1 + c  / cdr)
    phih = (phi + hdr * c / cdr) / (1 + c  / cdr)
    lamh = np.sqrt(k * D * ch)
    blam = b / lamh
    Lamh = 1 / (blam * np.cosh(blam) / np.sinh(blam) + (b /c) / (D / w))
    sinf = N * ch * (1 - Lamh) - Lamh * (hlr - phih)
    #print('N={:.4f} m/d, ch={:.2f} d, Lamh={:.3f} [-], hlr={:.2f} m, phih={:.2f} m, sinf={:.2f} '.format(
    #      N, ch, Lamh, hlr, phih, sinf))
    Th = mu * ch
    rising = (sinf + phih) > h0
    r = (h0  - phih - sinf) / (hdr - phih - sinf)
    thit = Th * np.log(r)
    return {'ch': ch, 'phih': phih, 'sinf':sinf, 'Th':Th,
        'rising':rising, 'r':r, 'thit':thit}

def headt(tau=None, N=None, b=None, mu=None, k=None, D=None, c=None, cdr=None,
          h0=None, phi=None, hdr=None, w=None, hlr=None, **kw):
    """Compute the head as a function of time since t0"""
    hatpars = get_hatpars(b=b, mu=mu, c=c, k=k, D=D, h0=h0, phi=phi,
                          hdr=hdr, cdr=cdr, w=w, hlr=hlr, N=N, **kw)
    
    ch, phih, sinf, Th, rising, r, thit = [hatpars[k]
                            for k in ['ch','phih','sinf','Th','rising','r','thit']]
    
    e = np.exp(- tau / Th)
    htau = phih + (h0 - phih) * e + sinf * (1 -e)
    dhdt = (-(h0 - phih) * e + sinf * e) / Th
    hinf = sinf + phih
    return htau, dhdt, thit, hinf, phih
```

# Show wat happens before and after the water table hits the drainage level

The result depends on the value of $c_{dr}$.

The smaller $c_dr$, the steeper the head after the hit, but also the lower the final head.

```python
kw = dict(data.iloc[0])

cdrain = kw['cdr']
kw.update(cdr=np.inf, h0=0.)

b, mu, k, D, c, w, N = [kw[k] for k in ['b','mu','k','D','c','w','N']]

ttl1 = 'head behavior after hiting drainage level for: '
ttl2 = 'b={:.0f} m, mu={:.3f} [-], k={:.0f} m/d, D={:.0f} m, c={:.4g} d, w={:.0f} d, N={:.3f} m/d'.format(
                                            b, mu, k, D, c, w, N)
ax = ga.newfig(ttl1 + ttl2, 'time [d]', 'head [m]', size_inches=(14, 8))

t  =  np.linspace(0., 30., 301)
t0 =  t[0]

# Simulation before the water table table hits the drainage level 
ht   , dhdt, thit, hinf, phih = headt(tau=t-t0, **kw)         # entire curve
hthit,    _,    _,    _,    _ = headt(tau=t[0] + thit, **kw)  # the hit (orange dot)

ax.plot(t,    ht,         label=', $c_{dr}=\infty$, ht')
ax.plot(thit, hthit, 'o', label='$ht=z_{dr}$, hit')

# Simulation after the hit, using differnt vaues for cdr
t0 = thit
for cdrain in [np.inf, 100, 30, 10, 3., 1, 0.3, 0.1]:
    kw.update(h0=hthit, cdr=cdrain)

    ht, dhdt, _, hinf, phih = headt(tau=t[t>=thit] - thit, **kw)
    hl = ax.plot(t[t>=thit], ht, label='ht, after hit, cdr={:.4g}'.format(cdrain))
    clr = hl[0].get_color()
    ax.hlines(hinf, t[0], t[-1], color=clr, linestyles='dashed', label='cdr={:.4g} $h_\infty$={:.2f} m'
              .format(cdr, hinf))
    #ax.hlines(phih, t[0], t[-1], colors=clr, linestyles='solid', label='phih={:.2f} m'.format(phih))

#ax.hlines(kw['hlr' ], t[0], t[-1], colors='cyan',   linestyles='dashed', label='hlr={:.2f} m'.format(kw['hlr']))
#ax.hlines(kw['phi' ], t[0], t[-1], colors='pink',   linestyles='dashed', label='$\phi$={:.2f} m'.format(kw['phi']))

ax.hlines(kw['hdr' ], t[0], t[-1], colors='orange', linestyles='dashed', label='zdr={:.2f} m'.format(kw['hdr']))

ax.legend()
plt.show()
     
```

## $h_\infty$ and $T_h \left(\frac{ d \overline h}{dt}\right)_{t=0}$ both as a function of the drainage resistance $c_{dr}$.

```python
kw = dict(data.iloc[0])

t0 =    0. # D At this time we want the rate of change
t1 = 1000. # D at this time (final time) we want the levels
cdrain = kw['cdr']

titles = ['head as function cdr at t={} d, zdr={:.2f} m, phi={:.2f} m, hlr={:.2f} m'
          .format(t1, kw['hdr'], kw['phi'], kw['hlr']),
         'head change rate at t={} d as function cdr, zdr={:.2f} m, phi={:.2f} m, hlr={:.2f} m'
          .format(t0, kw['hdr'], kw['phi'], kw['hlr'])]
          
ax = ga.newfig2(titles, 'cdr [d]', ['head [m]', 'change rate hd/dt [m/d]'], xscale='log', size_inches=(14, 8))

cdrs = np.logspace(-3, 5, 81) # set of cdr to use
ht   = np.zeros_like(cdrs)
dhdt = np.zeros_like(cdrs)
thit = np.zeros_like(cdrs)
hinf = np.zeros_like(cdrs)
phih = np.zeros_like(cdrs)
Th   = np.zeros_like(cdrs)

# We have four data sets, and so also four lines
clrs = ['b', 'r', 'g', 'k']
lss  =['solid', 'dashed', 'dashdot', 'dotted']

# For each case
for ip, clr in zip(data.index, clrs):
    kw = dict(data.iloc[ip])
    N = kw['N']    
    # For all cdr values
    for ic, cdr in enumerate(cdrs): 
        kw.update(cdr=cdr, h0=kw['hdr'])
        Th[ic] = dict(get_hatpars(**kw))['Th']
        ht[ic], _       , thit[ic], hinf[ic], phih[ic] = headt(t1, **kw)
        _     , dhdt[ic], _       , _       , _        = headt(t0, **kw)

    ax[0].plot(cdrs, ht,   clr, ls= 'solid', label='ht, N={}'.format(N))
    ax[1].plot(cdrs, Th * dhdt, clr, ls= 'solid', label='T * dhdt, N={}'.format(N))
    #ax[0].plot(cdrs, hinf, clr, marker= 'x', label='h(t={:.0f} d), N={} m/d'.format(t, N))
ax[0].plot(cdrs, phih, clr, ls='dashed', label='phih, N={} m/d')

ax[0].legend()
ax[1].legend()
```

As is clearly demonstrated here, mainly in the lower figure, that for certain combinations of $N$ and $c_{dr}$ the sign of the rate of change of the head at $t=0$ can change form positive to negative. It is clear because the sign for $\hat c=c$ i.e. for $c_dr=\infty$, that is on the right hand side of the picture may be negative while that at lower values of $c_{dr}$ it may be positive. This is never the case for $c_dr=0$, because then the rate of change at $t=0$ and even at $t>0$ will always be zero.

The cuves voor head at $t=\infty$ and of $d\overline h / dt$ look the exactly same. According to the theory, they indeed are completely similar, however they are shifted along the vertical axis relative to one another. We have
$overline h$ for $t \rightarrow \infty$ equals $\hat T\frac{\overline h}{dt} + \left(z_{dr} - \hat\phi\right)$ for $t=0$.
$$\overline h_{t=\infty} = \hat T \left(\frac{d \overline h}{dt}\right)_{t=0} + \left(z_{dr} - \hat\phi\right)$$

Also notice that $\hat\phi$ runs from $\phi=0$ to $z_{dr}=0.5$ when $c_{dr}$ declines from $\infty$ to 0.

```python
kw = dict(data.iloc[0])

t0 =    0. # D At this time we want the rate of change
t1 = 1000. # D at this time (final time) we want the levels

kw = dict(data.iloc[0])
b, mu, k, D, w, phi, hdr, hlr = [kw[k] for k in ['b', 'mu', 'k', 'D', 'w', 'phi', 'hdr', 'hlr']]

ttl2 = '$b$={:.0f} m, $\mu$={:.3f} [-], $k$={:.0f} m/d, $D$={:.0f} m, $\phi$={:.2f} m, $zdr$={:.2f} m, $hlr$={:.2f} m'.format(
                b, mu, k, D, w, phi, hdr, hlr)

ax = ga.newfig2(['$h_\infty}$ as a function of $c_{dr}$: '  + ttl2,
                 'dh/dt[t=0] as a function of $c_{dr}$: ' + ttl2],
                '$c_{dr}$ [d]', ['elevation [m]', 'dh/dt[0] [m/d]'], xscale='log')
ax2 = ga.newfig('$\Lambda$ as a function of $c_{dr}$: ' + ttl2, '$c_{dr}$ [d]', '$\hat\Lambda$ [-]',
                xscale='log', size_inches=(12, 4))

cdrs = np.logspace(-3, 5, 81) # set of cdr to use
ht   = np.zeros_like(cdrs)
dhdt = np.zeros_like(cdrs)
thit = np.zeros_like(cdrs)
hinf = np.zeros_like(cdrs)
phih = np.zeros_like(cdrs)
Th   = np.zeros_like(cdrs)
Lamh = np.zeros_like(cdrs)


# We have four data sets, and so also four lines
clrs = ['b', 'r', 'g', 'k']
lss  =['solid', 'dashed', 'dashdot', 'dotted']

# For each case
for ip, clr in zip(data.index, clrs):
    kw = dict(data.iloc[ip])
    N = kw['N']
    # For all cdr values
    for ic, cdr in enumerate(cdrs): 
        kw.update(cdr=cdr, h0=kw['hdr'])
        hatpars = dict(get_hatpars(**kw))
        ch, sinf[ic], Th[ic] = [hatpars[k] for k in ['ch', 'sinf', 'Th']]
        blamh = b / np.sqrt(k * D * ch)
        Lamh[ic] = 1 / (blamh * np.cosh(blamh) / np.sinh(blamh) + (b /ch) / (D / w))

        ht[ic], _       , thit[ic], hinf[ic], phih[ic] = headt(t1, **kw)
        _     , dhdt[ic], _       , _       , _        = headt(t0, **kw)

    ax[0].plot(cdrs, ht,   clr, ls= 'solid', label='ht, N={}'.format(N))
    ax[1].plot(cdrs, dhdt, clr, ls= 'solid', label='dhdt, N={}'.format(N))
    ax2.plot(cdrs,  Lamh, label='$\hat\Lambda$, N={:.4f} m/d'.format(N))
    ax[0].plot(cdrs, phih, clr, ls='dashed', label='phih, N={} m/d')

TOL = 0.1
ax[1].set_ylim((-0.1, 1))
ax[0].legend()
ax[1].legend()
ax2.legend()
```

Instead of $\left(\frac{d \overline h}{dt}\right)_{t=0}$ we now plotted $\left(T_h \frac{d \overline h}{dt}\right)_{t=0}$, which shows that the rate of change of the water table at $t=0$ grows with decline of $c_dr$. However, the curves also show that the sign of $\left(\frac{d \overline h}{dt}\right)_{t=0}$ depends on the value of $c_{dr}$. 

The last graph shows that $\Lambda\rightarrow 0$ for $c_{dr}\rightarrow 0$. 

## Simulatie met drainageweerstand $c_{dr}=0$

In this example, we simulate the head using $c_{dr}=\infty$ until the moment that the wate table hits the drainage level. At that moment we subtract $q_{dr}=\mu\left(\frac{\overline h}{dt}\right)_{\overline h = z_{dr}}$ from $N$. This is exactly the drainage for $c_{dr}=0$, and, hence, the water table will remain at the drainage level after the hit.

```python
kw = dict(data.iloc[0])
b, mu, k, D, c, w, N, phi = [kw[k] for k in ['b','mu','k','D','c','w','N', 'phi']]

ttl1 = 'Water table befofe and after hiting drainage level: '
ttl2 = 'b={:.0f} m, mu={:.3f} [-], k={:.0f} m/d, D={:.0f} m, c={:.4g} d, w={:.0f} d, N={:.3f} m/d'.format(
                                            b, mu, k, D, c, w, N)
ax = ga.newfig(ttl1 + ttl2, 'time [d]', 'head [m]', size_inches=(14, 8))

# Plotting until the water table hits the drainage level
t  =  np.linspace(0., 30., 301)
t0 =  t[0]
kw.update(cdr=np.inf, h0=0.)

ht   , dhdt, thit, hinf, phih = headt(tau=t-t0, **kw)
hthit,    _,    _,    _,    _ = headt(tau=t[0] + thit, **kw)

ax.plot(t,    ht,         label=', $c_{dr}=\infty, h_t$')
ax.plot(thit, hthit, 'o', label='$h_t=z_{dr}, h_{hit}$')

# Compute how much water we must drain to keep the water table at the drainge level
blam = b / np.sqrt(k * D * c)
Lam = 1 / (blam * np.cosh(blam) / np.sinh(blam) + (b / c) / (D / w))
q_dr = N - (hthit - phi) / c / (1 - Lam) + (hlr - phi) / c * Lam / (1 - Lam) 

kw.update(h0=hthit, N=N - q_dr) # subtract q_dr from N
ht, dhdt, _, hinf, phih = headt(tau=t[t>=thit] - thit, **kw)

ax.plot(t[t>=thit], ht, label='$h_t$ after the hit, $c_{dr}$ = 0')

ax.legend()
plt.show()     
```

This shows that by subtracting the precomputed $q_{dr}$ from $N$ when the water table hits the drainage level (orange point), then the computed water table will remain at the drainage level as the theory predicts.
