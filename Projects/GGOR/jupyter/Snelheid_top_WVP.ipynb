Vertikale stroming als gevolg van neerslagoverchot. Nul onderin op $z=0$, $-n$ bovenin op $z=H$:

$$ v_z = - \frac {n z} {H}$$


Horizontale snelheid naar onttrekking op $x=0$, met uniforme basisstroming evenwijdig aan de $x-as$

$$ v = -\frac Q {2 \pi D \epsilon r} - \frac {i k e_x} {\epsilon}$$

Dus op de $x$-as:

$$ v_x = -\frac Q {2 \pi D \epsilon x} - \frac {i k e_x} {\epsilon}$$

Stagnatipunt benedenstrooms:

$$ \frac Q {2 \pi x_s D \epsilon} = \frac {i k} \epsilon$$


$$ x_s = \frac Q {2 \pi D i k} $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Voorbeeld

Radiale stroming, getekend wordt de x-as waar de uniforme stroming is van links naar rechts. De onttrekking bevindt zich op $x=0$. Stagnatiepunt ligt op $x = -510$ m. De peiltjes zijn verticaal geschaald om ze zichtbaar te maken. 

```python
# Voorgbeeld:

# onttrekking op x=0
#

Q = 2400 # m3/d
H = 30 # m
k = 25 # m/d
n = 0.001 # m/d
i = 0.001 # m/m
eps = 0.35 # m3/m3
r0 = 0.25 # m

# stagnation point
xs = Q / (2 * np.pi * H * i * k)
print(f'stagnation point: x={xs:.2f} m benedenstrooms van de put op x=0')

x = np.hstack((np.linspace(-1000, -25, 40), np.linspace(25, 550, 25)))
z = np.linspace(0, H, 10)
X, Z = np.meshgrid(x, z)
X[X==0] = r0
Z=Z[::-1]

vx = -Q/(2 * np.pi * H * eps * X) - i * k  / eps
vz = -n * Z / H

print('\nPlaatje van de snelheidsvectoren in het verticale vlak door de x-as')

fig, ax = plt.subplots()
ax.set_title('Snelheidsveld (verticale snelheden geschaad om ze zichtbaar te maken)')
ax.set_xlabel('x [m] op x-ax')
ax.set_ylabel('z [m] boven basis')
fig.set_size_inches(14, 8)
#ax.quiver(X, Z, vx, vz)
ax.quiver(X, Z, vx, 1000 * vz)
plt.show()
```

## Laat de richting van de stroming zien aan top verzadigde zone

```python
fig, ax = plt.subplots()
fig.set_size_inches(14, 2)
ax.set_title('Richting van de grondwatersnelheid op verschillende punten van de x-as top verz. zone')
a = 500.
ax.plot(X[0], np.zeros_like(X[0]), '.')

for i in range(len(X[0])):
    ax.plot([X[0][i], X[0][i] + a * vx[0][i]], [0 * Z[0][i], 0 * Z[0][i] + a * vz[0][i]])
#ax.set_ylim((-120., 0))
#ax.axis('equal')
```

# Wat is de hoek van de snelheidscomponenten aan de top van de aquifer?

We kunnen de hoek in graden laten zien voor alle punten langs de x-as, waarbij -180 horizontaal in de richting van de negatieve x-as is en 0 graden horizontaal in de richting van de positieve x-as is.

```python
print('Hoek van snelheisvector, 90 = verticaal.\n', [f'{f:.0f}' for f in np.arctan2(vz[0], vx[0]) / np.pi * 180])
```

Het stagnatiepunt bevindt zich op locatie 22, dus waar de hoek tussen -155 en -37 graden is. Precies op het stagnatiepunt is de hoek -90, dat punt ligt daar tussen in. Links daarvan is de snelheid naar links gericht, rechts daarvan naar rechts tot de plek van de put (tussen punt 40 en 41). Daar slaat de hoek van 0 om naar -180 want rechts daarvan is de snelheid weer negatief. De hoeken zijn bijna 0 of -180 behalve vlak rond het stagnatiepunt. Dat betekent dat de snelheidsvectoren behalve daar nagenoeg vlak zijn. De horizontale snelheid is steeds veel groter dan de verticale.

Hier laat ik zien dat de horizontale snelheid veel groter is dan de verticale. Ik geef daarbij naast elkaar het puntnummer, de x van het punt, de hoek, de horizontale component en de verticale component. Dit voor de punten boven in het pakket.

```python
print('    {:5s} {:5s} {:5s}  {:5s} {:5s}'.format('nr', 'x', 'hoek', 'vx', 'vz'))

[f'{i:4.0f} {x:5.0f} {a:6.2f} {vx:6.3f} {vz:6.3f}' for i, x, a, vx, vz in np.vstack((np.arange(len(X[0])), X[0],
                                    np.arctan2(vz[0], vx[0]) * 180 / np.pi, vx[0], vz[0])).T]
```
